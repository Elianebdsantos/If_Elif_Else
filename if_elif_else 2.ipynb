```python
# float  indica número real (número inteiro ou decimal) input (espera receber algo, sempre deve ter um parenteses abrindo e fechando/Print tb tem que ter abertura do parenteses) int (numero inteiro)
#print recebendo como parametro uma variavel
#python é case sensitive
Nome= str(input("digite a nota"))
nota1= float(input("digite a nota")) #nota 1 vai receber um resultado
nota2= float (input("digite a nota"))
media=(nota1 + nota2)/2
if media >=7: #se a variavel média for maior ou igual a 7
  situação:"Aprovado"
  else:
    situação= "Reprovado"
print(f"A média do aluno é {media:.2f}") #(f"")significa concatenação / INTERPOLAÇÃO / no resultado para delimitar a casa depois da vírgula incluir .2f
print(f"Situação do aluno: {situação}")
print(f"Olá alundo {nome}, sua primeira nota foi {nota1} e sua segunda \n
f"nota foi {nota2}, sendo assim, sua média aritmetica foi de {media}, isso significa que você foi {situação} no seu curso CD")
```
