```python
#2. Faça um algoritmo que solicite o ano que o usuário nasceu, depois disso, faça o programa descrever se o usuário fará, se tem  ou já fez 18 anos.
ano_nascimento = int(input("Digite o ano que você nasceu: "))
ano= int(input("Digite o ano atual: "))
atual= ano - ano_nascimento
print(atual)
if atual >= 18:
  print("O usuário já tem mais que 18 anos")
elif atual == 18:
  print("O usuário fez 18 anos!")
else:
  print("Usuário ainda vai fazer 18 anos!")
```
